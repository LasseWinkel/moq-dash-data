```python
# Function to sample and smooth the data
def sample_and_smooth(data, timestamp_col, value_col, sample_rate=1, window_size=30):
    # Sampling
    sampled_data = data.iloc[::sample_rate, :]
    
    # Smoothing using a rolling window
    smoothed_values = sampled_data[value_col].rolling(window=window_size).mean()
    
    return sampled_data[timestamp_col], smoothed_values

# Sampling rate and window size for smoothing
sample_rate = 5  # Take every 30th frame
window_size = 1  # Rolling window size for smoothing

# Sample and smooth data for each scenario
timestamps_0, smoothed_0 = sample_and_smooth(df_data_0, '_1_rawVideoTimestamp', '_8_totalTime', sample_rate, window_size)
timestamps_1, smoothed_1 = sample_and_smooth(df_data_1, '_1_rawVideoTimestamp', '_8_totalTime', sample_rate, window_size)
timestamps_2, smoothed_2 = sample_and_smooth(df_data_2, '_1_rawVideoTimestamp', '_8_totalTime', sample_rate, window_size)
timestamps_3, smoothed_3 = sample_and_smooth(df_data_3, 'second', 'totalLatency', sample_rate, window_size)
timestamps_4, smoothed_4 = sample_and_smooth(df_data_4, 'second', 'totalLatency', sample_rate, window_size)
timestamps_5, smoothed_5 = sample_and_smooth(df_data_5, 'second', 'totalLatency', sample_rate, window_size)
```

```python
setups = ['MoQ Baseline Case', 'MoQ 10% Loss', 'MoQ 10 Mbps Bandwidth', 'LL-DASH Baseline Case', 'LL-DASH 5% Loss', 'LL-DASH 10 Mbps Bandwidth']

# MoQ data
encoding_times = [df_data_0['avg_encoding_time'].iloc[0], df_data_1['avg_encoding_time'].iloc[0], df_data_2['avg_encoding_time'].iloc[0], 0, 0, 0]  # in ms (MoQ only)
propagation_times = [df_data_0['avg_propagation_time'].iloc[0], df_data_1['avg_propagation_time'].iloc[0], df_data_2['avg_propagation_time'].iloc[0], 0, 0, 0]  # in ms (MoQ only)
decoding_times = [df_data_0['avg_decoding_time'].iloc[0], df_data_1['avg_decoding_time'].iloc[0], df_data_2['avg_decoding_time'].iloc[0], 0, 0, 0]  # in ms (MoQ only)

# LL-DASH data
time_behind_live = [0, 0, 0, df_data_3['avg_behindLive'].iloc[0], df_data_4['avg_behindLive'].iloc[0], df_data_5['avg_behindLive'].iloc[0]]  # in ms (LL-DASH only)
buffer_time = [0, 0, 0, df_data_3['avg_buffer'].iloc[0], df_data_4['avg_buffer'].iloc[0], df_data_5['avg_buffer'].iloc[0]]  # in ms (LL-DASH only)

# Set the position of the bars on the x-axis
x = np.arange(len(setups))

# Create the bar plot
fig, ax = plt.subplots(figsize=(12, 8), dpi=200)

# Plot each segment of the bar for MoQ
bar1 = ax.bar(x[:3], encoding_times[:3], label='Encoding Time (MoQ)', color='b')
bar2 = ax.bar(x[:3], propagation_times[:3], bottom=encoding_times[:3], label='Propagation Time (MoQ)', color='r')
bottom_decoding = np.add(encoding_times[:3], propagation_times[:3]).tolist()
bar3 = ax.bar(x[:3], decoding_times[:3], bottom=bottom_decoding, label='Decoding Time (MoQ)', color='g')

# Plot each segment of the bar for LL-DASH
bar4 = ax.bar(x[3:], time_behind_live[3:], label='Time Behind Live (LL-DASH)', color='purple')
bar5 = ax.bar(x[3:], buffer_time[3:], bottom=time_behind_live[3:], label='Buffer Time (LL-DASH)', color='orange')

# Add labels, title, and legend
ax.set_ylabel('Latency (ms)')
ax.set_xticks(x)
ax.set_xticklabels(setups, rotation=45, ha='right')

# Add horizontal dotted lines for 100ms, 200ms, 300ms, 400ms, and 500ms
for latency in [100, 200, 300, 400, 500]:
    ax.axhline(latency, color='gray', linestyle='dotted', linewidth=0.8)

# Display the plot
plt.legend(loc='upper left', ncol=2)


plt.savefig(f'plots/moq-dash-latencies.pdf', bbox_inches='tight')
```
