```python
import json
import pandas as pd
```

```python
# Open the JSON file
with open('base_data/publisherres1080fps30bit6gop(2,99,95)loss0delay50bw100.json', 'r') as file:
    json_data_data_0 = json.load(file)

# Convert JSON data to DataFrame
df_data_0 = pd.DataFrame(json_data_data_0)
```

```python
# Open the JSON file
with open('base_data/remotesubscriberres1080fps30bit6gop(2,99,95)loss0delay50bw100.json', 'r') as file:
    json_data_data_1 = json.load(file)

# Convert JSON data to DataFrame
df_data_1 = pd.DataFrame(json_data_data_1)
```

```python
# Selecting specific columns
selected_columns_0 = [
    '_0_frameId',
    '_1_rawVideoTimestamp',
    '_2_encodingTime',
    '_3_segmentationTimestamp',
    '_9_originalTimestampAttribute',
    '_10_encodedTimestampAttribute',
    '_13_sentBytes',
    '_15_sentType',
    '_19_segmentID'
]
selected_columns_1 = [
    '_0_frameId',
    '_5_receiveMp4FrameTimestamp',
    '_6_decodingTime',
    '_7_renderFrameTimestamp',
    '_11_decodedTimestampAttribute',
    '_12_renderTimestampAttribute',
    '_14_receivedBytes',
    '_16_receivedType',
    '_17_width',
    '_18_height'
]

df_selected_0 = df_data_0[selected_columns_0]
df_selected_1 = df_data_1[selected_columns_1]

df_merged = pd.merge(df_selected_0, df_selected_1, on='_0_frameId', how='inner')

# Creating the new columns
df_merged['_4_propagationTime'] = df_merged['_5_receiveMp4FrameTimestamp'] - df_merged['_3_segmentationTimestamp']
df_merged['_8_totalTime'] = df_merged['_7_renderFrameTimestamp'] - df_merged['_1_rawVideoTimestamp']

df_merged
```

```python
json_data = df_merged.to_json(orient='records')

# Save the JSON data to a file
output_file_path = 'res1080fps30bit6gop(2,99,95)loss0delay50bw100.json'
with open(output_file_path, 'w') as json_file:
    json_file.write(json_data)

print(f'JSON file saved to {output_file_path}')
```
